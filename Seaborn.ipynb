```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl

x = np.linspace(0,14,100)
for i in range(1,7):
    plt.plot(x,np.sin(x+i*0.5)*(10-i))
plt.show()
```
